```python
import seaborn as sns
import matplotlib.pyplot as plt

# Make sure win column exists
df["win"] = df["Closed PnL"] > 0

# Make sure classification column exists
df["classification"] = df["classification"].astype(str)

# Calculate win rate
win_rates = df.groupby("classification")["win"].mean().reset_index()

# Plot
plt.figure(figsize=(6,4))
sns.barplot(data=win_rates, x="classification", y="win")
plt.title("Win Rate: Fear vs Greed")
plt.xlabel("Market Sentiment")
plt.ylabel("Win Rate")
plt.show()
```
